```python
#T1 = r1*G1
xT1 = g1_scalar_base_mult(r1)
print("xT1 :")
for something in g1_marshall(xT1):
    print(something)
print("//---")
#T2 = r2*G1
xT2 = g1_scalar_base_mult(r2)
print("xT2 :")
for something in g1_marshall(xT2):
    print(something)
print("//---")
#T3 = r3*G1
xT3 = g1_scalar_base_mult(r3)
print("xT3 :")
for something in g1_marshall(xT3):
    print(something)
```

```python
#TP = T1 + T2 - T3
#xxT3 = curve_point(xT3.x, xT3.y.additive_inverse())
xxT3 = xT3.scalar_mul(gfp_1(order-1).value())
assert xxT3.is_on_curve() , " (xT3.x, -xT3.y) is not on curve "
xTP = g1_add(g1_add(xT1, xT2), xxT3)
print("xTP :")
for something in g1_marshall(xTP):
    print(something)
print("type(xTP) = {0}".format(type(xTP)))
```

```python
#T4 = r1*G2
xT4 = g2_scalar_base_mult(r1)
print("xT4 :")
for something in g2_marshall(xT4):
    print(something)
print("//---")
#T5 = r2*G2
xT5 = g2_scalar_base_mult(r2)
print("xT5 :")
for something in g2_marshall(xT5):
    print(something)
print("//---")
#T6 = r3*G2
xT6 = g2_scalar_base_mult(r3)
print("xT6 :")
for something in g2_marshall(xT6):
    print(something)
```

```python
#TQ = T4 + T5 - T6
xxT6 = xT6
xxT6.negate()
xTQ = g2_add(g2_add(xT4, xT5), xxT6)
print("xTQ :")
for something in g2_marshall(xTQ):
    print(something)
print("type(xTQ) = {0}".format(type(xTQ)))
```
