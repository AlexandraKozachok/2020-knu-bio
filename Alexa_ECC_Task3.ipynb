```python
def oneToGminusOneToinvG(xOne, G, invG):
    assert xOne in [-1,1], " def oneToG1minusOneToinvG1(xOne): argument type|range ERROR "
    if -1 == xOne:
        return invG
    else:
        return G
    
import random

def get_random_vector_of_minusOneAndOnes(xSize = 16):
    return [random.randrange(-1, 2, 2) for xIndex in range(xSize)]

def hamming_distance_typeList_length16_fast(xVec1, xVec2):
    return ((16 - sum([xVec1[i]*xVec2[i] for i in range(16)])) / 32)
```

```python
#3. Generate two vectors v1, v2 of [-1, 1] with length 16 
v1 = get_random_vector_of_minusOneAndOnes(16)
v2 = get_random_vector_of_minusOneAndOnes(16)
print("v1 = \n\t{0}".format(v1))
print("v2 = \n\t{0}".format(v2))
```

```python
#a. Transfer -1,1 to points. For v1 (registration template) you should use G2 generator, 
#      for v2 (authentication template) you should use G1 generator. 
#      Example:
#         v1 = -1, 1, -1, 1   v1points = invG2, G2, invG2, G2
#         v2 =  1, 1, -1, 1   v2points = G1, G1, invG1, G1
v1_as_points = [ oneToGminusOneToinvG(xElement, G2, invG2) for xElement in v1 ]
v2_as_points = [ oneToGminusOneToinvG(xElement, G1, invG1) for xElement in v2 ]
```

```python
#b. Generate array 's' random keys with length equal to bio template length (16)
#c. Generate array 't' random keys with length equal to bio template length (16)
#d. Generate array 'u' random keys with length equal to bio template length + 2 (18)
#e. Generate array 'v' random keys with length equal to bio template length + 2 (18)
s_vector = [ rand_elem() for i in range(16) ]
t_vector = [ rand_elem() for i in range(16) ]
u_vector = [ rand_elem() for i in range(18) ]
v_vector = [ rand_elem() for i in range(18) ]
```

```python
#f. Generate random h1 number and calculate point for G1 -> H1 (alread done)
#g. Generate random h2 number and calculate point for G2 -> H2 (alread done)
#j. Calculate gen1_h array keys: gen1_h[i] = s[i] * G1 + t[i] * H1
#i. Calculate gen2_h array keys: gen2_h[i] = u[i] * G2 + v[i] * H2
gen1_h = [ g1_add(G1.scalar_mul(s_vector[i]), H1.scalar_mul(t_vector[i])) for i in range(16) ]
gen2_h = [ g2_add(G2.scalar_mul(u_vector[i]), H2.scalar_mul(v_vector[i])) for i in range(18) ]
```

```python
#5. IPE: Registration:
#   Use presentation
#Decryption key generation (run on Device during registration)
# page 25

class Alexa_IPE_Registration:
    #Alexa_IPE_Registration(xMsk, v1_as_points, 16)
    
    def __init__(self, msk, c_vector, n):
        #self.msk = copy.deepcopy(msk)
        self.msk = msk
        self.c = c_vector
        self.n = n
        self.r0 = rand_elem()
        self.r0 = 1
        self.reg_template = []
        self.reg_template.append(G2.scalar_mul(self.r0))
        self.reg_template.append(H2.scalar_mul(self.r0))
        self.h2 = self.msk.h_g2
        self.reg_template += [xElement.scalar_mul(self.r0) for xElement in self.h2 ]
        #
        self.powers = self.msk.s_vector
        self.ProductTemp = \
        Alexa_IPE_Registration.akaProduct_g2(self.c, self.powers, 0, n, 0, 0)
        self.ProductTemp.negate()
        self.reg_template[2] = g2_add(self.reg_template[2], self.ProductTemp)
        #
        self.powers = self.msk.t_vector
        self.ProductTemp = \
        Alexa_IPE_Registration.akaProduct_g2(self.c, self.powers, 0, n, 0, 0)
        self.ProductTemp.negate()
        self.reg_template[3] = g2_add(self.reg_template[3], self.ProductTemp)
        #
        for i in range(0, 16, 1):
            self.reg_template[i+4] = g2_add(self.reg_template[i+4], self.c[i])
            
    @staticmethod            
    def akaProduct_g2(c_vector, power_vector, j, n, shift_c_index = 0, shift_power_index = 0):
        xAkaProduct = c_vector[j + shift_c_index].scalar_mul(gfp_1(power_vector[j + shift_power_index]).value())
        for i in range(j, n, 1):
            temp = c_vector[i + shift_c_index].scalar_mul(gfp_1(power_vector[i + shift_power_index]).value())
            xAkaProduct = g2_add(xAkaProduct, temp)
            del temp
        return xAkaProduct
```

```python
#6. IPE: Authentication:
#   Use presentation

class Alexa_IPE_Authentication:
    
    def __init__(self, msk, c_vector, n):
        #self.msk = copy.deepcopy(msk)
        self.msk = msk
        self.c = c_vector
        self.n = n
        self.r0 = rand_elem()
        self.r0 = gfp_1(order-1).value()
        self.r0 = 1
        self.auth_template = []
        self.G1zero = G1.scalar_mul(0)
        self.auth_template.append(self.G1zero)
        self.auth_template.append(self.G1zero)
        self.auth_template.append(G1.scalar_mul(gfp_1(self.r0).value()))
        self.auth_template.append(H1.scalar_mul(gfp_1(self.r0).value()))
        #
        self.h1 = self.msk.h_g1
        self.auth_template += [g1_add(self.h1[i].scalar_mul(gfp_1(self.r0).value()), self.c[i]) for i in range(16) ]
        #
        self.powers = self.msk.u_vector
        self.ProductTemp = \
        Alexa_IPE_Authentication.akaProduct_g1(self.auth_template, self.powers, 0, (self.n + 2), 2, 0)
        self.auth_template[0] = g1_add(self.auth_template[0], self.ProductTemp.scalar_mul(gfp_1(order-1).value()))
        #
        self.powers = self.msk.v_vector
        self.ProductTemp = \
        Alexa_IPE_Authentication.akaProduct_g1(self.auth_template, self.powers, 0, (self.n + 2), 2, 0)
        self.auth_template[1] = g1_add(self.auth_template[1], self.ProductTemp.scalar_mul(gfp_1(order-1).value()))

    @staticmethod            
    def akaProduct_g1(c_vector, power_vector, j, n, shift_c_index = 0, shift_power_index = 0):
        xAkaProduct = c_vector[j + shift_c_index].scalar_mul(gfp_1(power_vector[j + shift_power_index]).value())
        for i in range(j,n, 1):
            temp = c_vector[i + shift_c_index].scalar_mul(gfp_1(power_vector[i + shift_power_index]).value())
            xAkaProduct = g1_add(xAkaProduct, temp)
            del temp
        return xAkaProduct
```
